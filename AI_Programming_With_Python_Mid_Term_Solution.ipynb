```python
import numpy as np
import pandas as pd

np.random.seed(24)

num_employees = 50  # you can increase this if needed

# Employee data
emp_ids = np.arange(101, 101 + num_employees)
emp_ages = np.random.randint(20, 55, num_employees)
emp_income = np.random.randint(35000, 95000, num_employees)

# Fixed set of roles (repetition allowed)
roles_list = [
    "HR Manager",
    "Software Developer",
    "Accountant",
    "Marketing Executive",
    "Sales Executive",
    "Backend Developer",
    "HR Executive",
    "Financial Analyst",
    "Digital Marketer",
    "Operations Manager"
]

emp_roles = np.random.choice(roles_list, num_employees)

# Create DataFrame
data = pd.DataFrame({
    "emp_id": emp_ids,
    "emp_age": emp_ages,
    "emp_salary": emp_income,
    "role": emp_roles
})

# Save dataset
data.to_csv("staff_data.csv", index=False)
```

**Q2.**

```python
max_salary = np.max(emp_income)
min_salary = np.min(emp_income)
ave_salary = np.mean(emp_income)
ave_age = np.mean(emp_ages)

print(f"Maximum Salary - {max_salary}")
print(f"Minimum Salary - {min_salary}")
print(f"Average Salary - {ave_salary}")
print(f"Average Age - {ave_age}")
```

**Q3.**

```python
employees_by_ages = emp_ages[(emp_ages >= 25) & (emp_ages <= 40)]
employees_by_salary = emp_income[emp_income > 50000]
print(f"Employees through age - {employees_by_ages}")
print(f"Total count - {employees_by_ages.size}")
print("---------------------------------------------------------")
print(f"Employees through income - {employees_by_salary}")
print(f"Total count - {employees_by_salary.size}")
```

**Q4.**

```python
adjusted_arr = np.where(emp_ages < ave_age, emp_ages + 4, emp_ages + 2)

print(f"Original arrays - {emp_ages}")
print(f"Adjusted arrays - {adjusted_arr}")
print(f"Average Age - {ave_age}")
```
